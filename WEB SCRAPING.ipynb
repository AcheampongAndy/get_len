```python
import requests as rq
```

```python
url = "https://www.indeed.com/jobs?q=data%20science&l=Remote&vjk=a2c008533625ab9d"
```

```python
res = rq.get(url)
```

```python
print(res)
```

```python
from bs4 import BeautifulSoup as bs #importing the libraries
jobs = bs(res.content) #get content from page
#print(jobs)
#get element by id
results = jobs.find(id = "resultsCol") #we use id to find 
results
```

```python
#get element by class
c_resuts = results.find_all("div", class_="messageContainer")
c_resuts
```

```python
c_results1 = results.find_all("h2", class_="jobTitle")
job_titles = [l.text.strip() for l in c_results1] #get all job roles
job_titles
```

```python
import pandas as pd
df = pd.DataFrame(job_titles, columns = ["Job Title"])#put the data in a dataframe
df.head()
```

```python
df.tail()
```

```python
# collect our data into a csv
df.to_csv("Indeed Romote Jobs First Page.csv")
```
